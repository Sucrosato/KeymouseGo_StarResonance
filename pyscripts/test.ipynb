```python
import pyautogui
from time import sleep
```

```python
def get_cocoon_marker_pos(delay=0):
    sleep(delay)
    location = pyautogui.locateOnScreen(
                    '../data/pics/cocoon_marker.png',
                    region=[63, 40, 240, 240],
                    confidence=0.7,
                    grayscale=True
                )
    point = pyautogui.center(location)
    return [int(point.x), int(point.y)]
```

```python
zero_pos = get_cocoon_marker_pos()
zero_pos
```

```python
get_cocoon_marker_pos()
```

```python
def move_to_cocoon(pos):
    offset_x = pos[0] - 138.5
    offset_y = pos[1] - 169
    if offset_x < 0:
        pyautogui.keyDown('a')
        sleep(max(min(abs(offset_x) /8), 0))
        pyautogui.keyUp('a')
    elif offset_x > 0:
        pyautogui.keyDown('d')
        sleep(max(min(offset_x /8), 0))
        pyautogui.keyUp('d')
    sleep(0.5)
    if offset_y < 0:
        pyautogui.keyDown('w')
        sleep(max(min(abs(offset_y) /8), 0))
        pyautogui.keyUp('w')
    elif offset_y > 0:
        pyautogui.keyDown('s')
        sleep(max(min(offset_y /8), 0))
        pyautogui.keyUp('s')
    
```

```python
move_to_cocoon(get_cocoon_marker_pos(1))
```

```python
import pandas as pd

# 1. 读取 CSV 文件
df = pd.read_csv(r'E:\Apps\git_repos\KeymouseGo\dist\data\daily_balance_report.csv')

# 2. 将所有浮点数转换为整数
# 注意：如果某列包含空值 (NaN)，直接转 int 会报错，Pandas 默认会将含空值的列转为 float
# 我们通常先填充空值或者只转换数值列
cols_to_convert = df.columns[1:]

# 3. 仅转换这些列：填充空值为0并转为整数
df[cols_to_convert] = df[cols_to_convert].fillna(0).astype(int)
df.to_csv('output_pandas.csv', index=False)
print("转换完成！")
```
